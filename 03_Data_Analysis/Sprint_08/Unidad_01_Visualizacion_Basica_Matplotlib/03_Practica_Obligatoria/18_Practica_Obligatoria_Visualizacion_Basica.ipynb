#### #1.2  

Recrea la siguiente figura, para ello construye en apartados separados cada grfico y luego haz un apartado en el que se junten los cuatro.  

Las figuras son:
1. Las ciudades ms pobladas, con los numeros de las poblaciones (tendrs que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrn, con trasparencia)  

2. Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand ITC", tamao 18, color azul acero o "steelblue")
3. El scatter de rea de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python
# Ciudades ms pobladas (Diagrama de barras)

top_cities = cities_clean.nlargest(5, "population_total")

fig1, ax1 = plt.subplots()
#Aqui la selecion de colores:
colores = ["#7777FF", "#BDBDBD", "#FF8888", "#BDBE8C", "#D99494"]

bars = ax1.bar(top_cities['city'], top_cities['population_total']/1000, color=colores)
ax1.set_title("Ciudades ms pobladas de California")
ax1.set_ylabel("Habitantes (miles)")
ax1.set_xlabel("Ciudades")

#El bucle recorre cada grafico y escribe encima su valor
for bar, value in zip(bars, top_cities['population_total']/1000):
    ax1.text(bar.get_x() + bar.get_width()/2,
             bar.get_height() + 5,
             f"{value:.3f}",
             ha='center', va='bottom')

# Guardar figura 1
fig1.savefig("./img/Diagrma_Barras.png")
```

```python
# Histograma del rea total (INDIVIDUAL)

fig2, ax2 = plt.subplots()

ax2.hist(cities_clean['area_total_km2'], bins=50, color='steelblue', label="Hist Area")

# Estilo Bradley Hand ITC para texto general
estilo_ejes = {
    'fontsize': 18,
    'color': 'black',
    'weight': 'normal',
    'family': 'Bradley Hand ITC', #El tipo de letra que dice el enunciado
    'style': 'italic'
}

ax2.set_xlabel(r"Superficie ($\mathit{km}^{2}$)", fontdict=estilo_ejes)# Aqui utilizo Latex
ax2.set_ylabel("Nmero de ciudades", fontdict=estilo_ejes)

ax2.set_title("Histograma del rea total")
ax2.legend()

# Guardar figura 2
fig2.savefig("./img/fig_histograma.png")
```

```python
# Scatter individual con flechas

fig3, ax3 = plt.subplots()

x = cities_clean['area_water_km2']
y = cities_clean['population_total'] / 1000

ax3.scatter(x, y, alpha=0.7, color='green')
ax3.set_title("Poblacin por kilmetro cuadrado de agua")
ax3.set_xlabel("rea de agua en la ciudad ($km^2$)")
ax3.set_ylabel("Poblacin (miles habitantes)")

#Coordenadas para mover el primer punto el de mas arriba
punto1_x = 87
punto1_y = 3850
#Coordenadas para mover el segundo punto 
punto2_x = 120
punto2_y = 1400
#Coordenada para mover el texto
texto_x = 70
texto_y = 3400


ax3.text(texto_x-10, texto_y-4, "Outliers", fontsize=10, color="black", weight="normal")

props_flecha = {
    "arrowstyle": "->", #estilo de flecha
    "color": "black",   #Color 
    "linewidth": 1      #Grosor de la linea
}

ax3.annotate("", xy=(punto1_x, punto1_y), xytext=(texto_x, texto_y+180), arrowprops=props_flecha)
ax3.annotate("", xy=(punto2_x, punto2_y), xytext=(texto_x, texto_y-100), arrowprops=props_flecha)

# Guardar figura 3
fig3.savefig("./img/fig_scatter.png")
```

```python
# Clculo de estadsticas del boxplot

altitudes = cities_clean['elevation_m']

q1 = altitudes.quantile(0.25)
q2 = altitudes.median()
q3 = altitudes.quantile(0.75)
iqr = q3 - q1

lim_inf = q1 - 1.5 * iqr
lim_sup = q3 + 1.5 * iqr

print("Q1:", q1)
print("Mediana:", q2)
print("Q3:", q3)
print("IQR:", iqr)
print("Bigote inferior:", lim_inf)
print("Bigote superior:", lim_sup)

# Boxplot de altitud (horizontal)

fig4, ax4 = plt.subplots(figsize=(8, 2.5))

ax4.boxplot(
    cities_clean['elevation_m'],
    vert=False,
    patch_artist=True,
    boxprops=dict(facecolor='red', color='black'),
    whiskerprops=dict(color='black'),
    capprops=dict(color='black'),
    medianprops=dict(color='black'),
    flierprops=dict(marker='o', color='black', alpha=0.5)
)

ax4.set_title("Altitud de las ciudades de California")
ax4.set_xlabel("Altitud (m)")
ax4.set_yticks([1])
ax4.set_yticklabels(["1"])

# Guardar figura 4
fig4.savefig("./img/fig_boxplot.png")
```

```python
# Figura conjunta 2x2

'''Aqui 'fig' es un objeto figura, 'axs' es un array de objetos axes y plt.subplots(2,2) le estoy diciendo crea una figura con 4 subplots (dos filas y dos columnas) es como una matriz de ejes'''
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Subplot 1: Barras (Aqui es la posicion de arriba a la izquierda)
axs[0, 0].bar(top_cities['city'], top_cities['population_total']/1000, color=colores)
axs[0, 0].set_title("Ciudades ms pobladas")
axs[0, 0].set_ylabel("Habitantes (miles)")
axs[0, 0].set_xlabel("Ciudades")

for bar, value in zip(axs[0, 0].patches, top_cities['population_total']/1000):
    axs[0, 0].text(bar.get_x() + bar.get_width()/2,
                   bar.get_height() + 5,
                   f"{value:.1f}",
                   ha='center', va='bottom')

# Subplot 2: Histograma (Aqui es la posicion de arriba a la derecha)
axs[0, 1].hist(cities_clean['area_total_km2'], bins=50, color='steelblue', label="Hist Area")
axs[0, 1].set_title("Histograma del rea total")
axs[0, 1].set_xlabel(r"Superficie ($\mathit{km}^{2}$)", fontdict=estilo_ejes)#Aqui utilizo latex 
axs[0, 1].set_ylabel("Nmero de ciudades", fontdict=estilo_ejes)
axs[0, 1].legend()

# Subplot 3: Scatter (Aqui es la posiicon de abajo a la izquierda)
axs[1, 0].scatter(x, y, alpha=0.7, color='green')
axs[1, 0].set_title("Poblacin por kilmetro cuadrado de agua")
axs[1, 0].set_xlabel("rea de agua en la ciudad ($km^2$)")
axs[1, 0].set_ylabel("Poblacin (miles habitantes)")
axs[1, 0].text(texto_x-10, texto_y-4, "Outliers", fontsize=10, color="black", weight="normal")
axs[1, 0].annotate("", xy=(punto1_x, punto1_y), xytext=(texto_x, texto_y+180), arrowprops=props_flecha)
axs[1, 0].annotate("", xy=(punto2_x, punto2_y), xytext=(texto_x, texto_y-100), arrowprops=props_flecha)

# Subplot 4: Boxplot (aqui es la posicon de abajo a la derecha)
axs[1, 1].boxplot(
    cities_clean['elevation_m'],
    vert=False,
    patch_artist=True,
    boxprops=dict(facecolor='red', color='black'),
    whiskerprops=dict(color='black'),
    capprops=dict(color='black'),
    medianprops=dict(color='black'),
    flierprops=dict(marker='o', color='black', alpha=0.5)
)
axs[1, 1].set_title("Altitud de las ciudades")
axs[1, 1].set_xlabel("Altitud (m)")
axs[1, 1].set_yticks([1])
axs[1, 1].set_yticklabels(["1"])

plt.tight_layout()

# Guardar figura conjunta
fig.savefig("./img/fig_conjunta.png")

plt.show()
```
