### #1.2

Crea una funcin similar a la que se pide en el ejercicio 6 de la segunda tanda de ejercicios del workout. Es decir, una funcin para pintar diagramas de dispersion que tenga seis argumentos, pero ojo al cambio en el quinto:
1. Un dataframe
2. Una columna numrica del dataframe
3. Una columna numrica del dataframe diferente a la anterior.
4. Un columna categrica (Opcional, valor por defecto None)
5. Un valor para el tamao de los puntos o una cadena, pero que puede ser una cadena de caracteres, en cuyo caso se debe interpretar como una de las columnas del dataframe cuyos valores deben controlar el tamao
6. Un sexto argumento "escala", que permita manipular la escala del tamao en el caso de que el quinto argumento sea una cadena. Este argumento multiplicar a los valores de la columna indicada en el parmetro anterior (as si quiero dividir por 10000, este sexto argumento tendr que valer 1/10000)

La funcin debe pintar el diagrama de dispersin de la columna numrica 1, en el eje X, frente a la columna numrica 2, en el eje Y, usando un cdigo de colores en funcin de la columna categrica y con el tamao de los puntos segn el argumento correspondiente. 

Si no se diera una columna categrica como argumento, el color no debe pasarse como parmetro a la funcin de Seaborn o Matplotlib que uses.

EXTRA: Aade un argumento que controle si la funcin pinta una leyenda para colores y tamaos o un colorbar y una leyenda para los tamaos. NO TE PREOCUPES POR EL FORMATO DE LA LEYENDA (lo veremos en la sesin en vivo)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def plot_scatter(df, col_x, col_y, col_cat=None, size=50, escala=1, legend_type='legend'):
    """
    La funcion crea diagrama de dispersion con opciones de color por categoria y tamao de la variable
    
    Parmetros:
    -----------
    df : DataFrame
        El dataframe con los datos
    col_x : str
        Nombre de la columna numrica para el eje X
    col_y : str
        Nombre de la columna numrica para el eje Y
    col_cat : str, opcional (default=None)
        Nombre de la columna categrica para el color
    size : int o str (default=50)
        Si es int: tamao fijo de los puntos
        Si es str: nombre de la columna que controla el tamao
    escala : float (default=1)
        Factor de escala para multiplicar los tamaos cuando size es una columna
    legend_type : str (default='legend')
        'legend': muestra leyendas para colores y tamaos
        'colorbar': muestra colorbar para colores y leyenda para tamaos
    """
    
    # Crear figura
    plt.figure(figsize=(10, 6))
    
    # Determinar si size es una columna o un valor fijo
    if isinstance(size, str):
        # size es una columna del dataframe
        sizes = df[size] * escala
    else:
        # size es un valor fijo
        sizes = size
    
    # Pintar el grfico segn si hay columna categrica o no
    if col_cat is None:
        # Sin columna categrica, no usar color
        if isinstance(size, str):
            plt.scatter(df[col_x], df[col_y], s=sizes, alpha=0.6, edgecolors='black', linewidth=0.5)
        else:
            plt.scatter(df[col_x], df[col_y], s=sizes, alpha=0.6, edgecolors='black', linewidth=0.5)
    else:
        # Con columna categrica, usar colores
        if legend_type == 'colorbar':
            # Convertir categoras a nmeros para colorbar
            categories = df[col_cat].astype('category')
            colors = categories.cat.codes
            scatter = plt.scatter(df[col_x], df[col_y], c=colors, s=sizes, 
                                 alpha=0.6, cmap='viridis', edgecolors='black', linewidth=0.5)
            
            # Aadir colorbar
            cbar = plt.colorbar(scatter)
            cbar.set_label(col_cat)
            # Ajustar etiquetas del colorbar a las categoras originales
            cbar.set_ticks(range(len(categories.cat.categories)))
            cbar.set_ticklabels(categories.cat.categories)
        else:
            # Usar seaborn para leyenda automtica de colores
            if isinstance(size, str):
                sns.scatterplot(data=df, x=col_x, y=col_y, hue=col_cat, 
                               size=size, sizes=(20, 200), alpha=0.6, 
                               edgecolor='black', linewidth=0.5)
            else:
                sns.scatterplot(data=df, x=col_x, y=col_y, hue=col_cat, 
                               s=sizes, alpha=0.6, edgecolor='black', linewidth=0.5)
    
    # Aadir etiquetas
    plt.xlabel(col_x)
    plt.ylabel(col_y)
    plt.title(f'{col_y} vs {col_x}')
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
```

- Aqui lo hago con una funcion

```python
import numpy as np
import matplotlib.pyplot as plt

def scatter_geo(
    df,
    col_x,
    col_y,
    size_col,
    color_col,
    escala=1,
    titulo="Diagrama de dispersin geogrfico"
):
    """
    df: DataFrame
    col_x: columna numrica para eje X
    col_y: columna numrica para eje Y
    size_col: columna numrica para tamao de puntos
    color_col: columna numrica para color
    escala: factor multiplicativo para el tamao
    titulo: ttulo del grfico
    """

    # Tamao escalado
    sizes = df[size_col] * escala

    # Crear figura
    plt.figure(figsize=(20, 10))

    # Scatter principal
    scatter = plt.scatter(
        df[col_x],
        df[col_y],
        s=sizes,
        c=df[color_col],
        cmap="viridis",
        alpha=0.5,
        linewidth=0
    )

    # Etiquetas
    plt.xlabel(col_x)
    plt.ylabel(col_y)

    # Colorbar
    cbar = plt.colorbar(scatter)
    cbar.set_label(color_col)

    # Escala fija opcional (como en tu ejemplo)
    plt.clim(df[color_col].min(), df[color_col].max())

    # LEYENDA 2 FILAS  3 COLUMNAS
    valores_leyenda = [100, 200, 300, 400, 500, 600]

    for a in valores_leyenda:
        plt.scatter([], [], c="gray", alpha=0.4, s=a, label=f"{a} km$^2$")

    plt.legend(
        title=size_col,
        frameon=False,
        loc="upper right",
        ncol=3,              # tres columnas
        columnspacing=1.5,   # espacio entre columnas
        labelspacing=1.2,    # espacio entre filas
        scatterpoints=1
    )

    # Ttulo
    plt.title(titulo)

    plt.tight_layout()
    plt.show()
```

```python
cities_clean["log_poblacion"] = np.log10(cities_clean["population_total"])

scatter_geo(
    df=cities_clean,
    col_x="longd",
    col_y="latd",
    size_col="area_total_km2",
    color_col="log_poblacion",
    escala=1,
    titulo="California Cities: rea y Poblacin"
)
```

- Aqui utilizo la funcion de Jaime para crearme una grafica para ver en terminos relativos de que grupo hubo mas supervivientes de mujeres, nios y hombres.

```python
def plot_categorical_relationship_fin(df, cat_col1, cat_col2, relative_freq=False, show_values=False, size_group = 5):
    # Prepara los datos
    count_data = df.groupby([cat_col1, cat_col2]).size().reset_index(name='count')
    total_counts = df[cat_col1].value_counts()
    
    # Convierte a frecuencias relativas si se solicita
    if relative_freq:
        count_data['count'] = count_data.apply(lambda x: x['count'] / total_counts[x[cat_col1]], axis=1)

    # Si hay ms de size_group categoras en cat_col1, las divide en grupos de size_group
    unique_categories = df[cat_col1].unique()
    if len(unique_categories) > size_group:
        num_plots = int(np.ceil(len(unique_categories) / size_group))

        for i in range(num_plots):
            # Selecciona un subconjunto de categoras para cada grfico
            categories_subset = unique_categories[i * size_group:(i + 1) * size_group]
            data_subset = count_data[count_data[cat_col1].isin(categories_subset)]

            # Crea el grfico
            plt.figure(figsize=(10, 6))
            ax = sns.barplot(x=cat_col1, y='count', hue=cat_col2, data=data_subset, order=categories_subset)

            # Aade ttulos y etiquetas
            plt.title(f'Relacin entre {cat_col1} y {cat_col2} - Grupo {i + 1}')
            plt.xlabel(cat_col1)
            plt.ylabel('Frecuencia' if relative_freq else 'Conteo')
            plt.xticks(rotation=45)

            # Mostrar valores en el grfico
            if show_values:
                for p in ax.patches:
                    ax.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                                ha='center', va='center', fontsize=10, color='black', xytext=(0, size_group),
                                textcoords='offset points')

            # Muestra el grfico
            plt.show()
    else:
        # Crea el grfico para menos de size_group categoras
        plt.figure(figsize=(10, 6))
        ax = sns.barplot(x=cat_col1, y='count', hue=cat_col2, data=count_data)

        # Aade ttulos y etiquetas
        plt.title(f'Relacin entre {cat_col1} y {cat_col2}')
        plt.xlabel(cat_col1)
        plt.ylabel('Frecuencia' if relative_freq else 'Conteo')
        plt.xticks(rotation=45)

        # Mostrar valores en el grfico
        if show_values:
            for p in ax.patches:
                ax.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                            ha='center', va='center', fontsize=10, color='black', xytext=(0, size_group),
                            textcoords='offset points')

        # Muestra el grfico
        plt.show()
```

```python
plot_categorical_relationship_fin(titanic,"who", "alive", relative_freq=True, show_values=True)
```

Si, en trminos relativos en porcentaje de supervivientes de mujeres y nios es muy superior al de los hombres.

Orden de Supervivencia:

1) Mujeres (76%)
2) Nios (59%)
3) Hombres (16%)


```python
def tricategoricalmodify_analysis(df, directora, otras, relativa = False, muestra_valores = False):
    df = df.copy()

    col_directora = directora
    col_1 = otras[0]   # alive
    col_2 = otras[1]   # who

    # 1. Traducir yes/no  Superviviente/Fallecido
    if col_1 == "alive":
        df["alive_es"] = df["alive"].map({
            "yes": "Superviviente",
            "no": "Fallecido",
        })

        #  2. Convertir en categora ORDENADA
        #    Esto FIJA los colores SIEMPRE:
        #    - Fallecido  AZUL
        #    - Superviviente  NARANJA
        df["alive_es"] = pd.Categorical(
            df["alive_es"],
            categories=["Fallecido", "Superviviente"],   #  ORDEN CORRECTO
            ordered=True
        )

        col_1 = "alive_es"

    # 3. Dividir por clase
    diccionario_multivariante = {}
    for valor in df[col_directora].unique():
        diccionario_multivariante[valor] = df.loc[df[col_directora] == valor, [col_2, col_1]]

    # 4. Llamar a tu funcin SIN tocarla
    for valor, df_datos in diccionario_multivariante.items():
        print(f"\nRespuesta {valor}:")
        plot_categorical_relationship_fin(
            df_datos,
            col_2,
            col_1,
            relative_freq=relativa,
            show_values=muestra_valores
        )
        plt.show()
```

```python
tricategoricalmodify_analysis(
    titanic_copy,
    "class",
    ["alive", "who"],
    relativa=True,
    muestra_valores=True
)
```
